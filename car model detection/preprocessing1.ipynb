```python
import numpy as np 
import re
import os
import tensorflow as tf 
from tensorflow import keras
import shutil
```

```python
def create_folder(x):
    if os.path.exists(x):
        print("Folder already exists !!!")

    else:
        os.mkdir(x)
        print("{x} created Successfully !!!")
```

Find the names of cars

```python
unique_names = set()
directory = "D:\Datasets\car_make\images\\test"
for filename in os.listdir(directory):
    if filename.lower().endswith((('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'))):
        x = filename.split("_")[1]
        x = os.path.splitext(x)[0]
        unique_names.add(x)

print(unique_names)
```

Making Directories for the following name

```python
unique_names = list(unique_names)
for names in unique_names:
    create_folder(f"D:\Datasets\cars_new\\test\\{names}")
```

Transferring the images to their respective folders

```python
directories = []
new_path = "D:\Datasets\cars_new\\test"
for entry in os.listdir(new_path):
    entry_path = os.path.join(new_path,entry)
    if os.path.isdir(entry_path):
        directories.append(entry_path)

car_names = []

for i in directories:
    i = i.split("\\")
    car_names.append(i[-1])

print(car_names)
```

Function to get the index

```python
def get_index(lst,item):
    for i, name in enumerate(lst):
        if name == item:
            return i
    return None
```

```python
for cars in os.listdir(directory):
    crs = cars.split("_")[1]
    c_name = crs[:-4]
    if c_name in car_names:
        index = get_index(car_names,c_name)
        source_path = os.path.join(directory,cars)
        destination_path = os.path.join(new_path, car_names[index])
        shutil.copy(source_path,destination_path)
        print(f"{destination_path} copied Successfully...")
```
